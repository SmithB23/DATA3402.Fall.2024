```python
class counter:
    def __init__(self,max):
        self.count=0
        self.max=max
    
    def counting(self):
        if self.count<=self.max:
            self.count=self.count + 1
            return (self.count)
        else:
            return ("You are trying to increment beyond the maximum value")
    def reset(self):
        self.count=0
```
